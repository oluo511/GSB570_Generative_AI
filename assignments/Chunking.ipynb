## Cleaning Data

```python
f1.isna().sum()
```

```python
f1['AvgPitStopTime'] = f1['AvgPitStopTime'].fillna(f1['AvgPitStopTime'].mean())
f1['Lap Time Variation'] = f1['Lap Time Variation'].fillna(0)
f1['Tire Usage Aggression'] = f1['Tire Usage Aggression'].fillna(0)
f1['Fast Lap Attempts'] = f1['Fast Lap Attempts'].fillna(0).round(0).astype(int)
f1['Driver Aggression Score'] = f1['Driver Aggression Score'].fillna(0).round(2)

f1['Stint'] = f1['Stint'].fillna('unknown')
f1['Tire Compound'] = f1['Tire Compound'].fillna('UNKNOWN')
f1['Stint Length'] = f1['Stint Length'].fillna(0).astype(int)

f1['Pit_Lap'] = f1['Pit_Lap'].fillna('unknown')
f1['Pit_Time'] = f1['Pit_Time'].fillna('unknown')
```

```python
f1[f1['Race Name'].isnull()].head(10)
```

```python
f1_clean = f1.dropna(subset=['Race Name', 'Date', 'Location', 'Country'])
```

## Write summaries for races

```python
def generate_summary(row):
    # Start of the summary: general race and driver info
    summary = (
        f"In the {row['Season']} Formula 1 season at {row['Race Name']} "
        f"({row['Location']}, {row['Country']}), held on {row['Date']}, "
        f"{row['Driver']} drove for {row['Constructor']}. They completed "
        f"{int(row['Laps'])} laps and finished in position {int(row['Position'])} overall. "
    )

    # Handle stint data more naturally
    if row['Stint'] == 'unknown' or row['Tire Compound'] == 'UNKNOWN':
        summary += "Stint information was unavailable for this race. "
    else:
        summary += (
            f"During stint number {row['Stint']}, they used {row['Tire Compound']} tires for "
            f"{int(row['Stint Length'])} laps, pitting on lap {row['Pit_Lap']} with a pit stop time of "
            f"{row['Pit_Time']} seconds. "
        )

    # Pit stops, lap attempts, and position changes
    summary += (
        f"The driver made {int(row['TotalPitStops'])} pit stop(s) total during the race, with an average pit stop time of "
        f"{round(row['AvgPitStopTime'], 2)} seconds. "
        f"Fast lap attempts during the race: {row['Fast Lap Attempts']}. "
        f"The driver changed position {round(row['Position Changes'], 1)} times. "
    )

    # Aggression and lap time data
    summary += (
        f"Aggression score for this stint was {round(row['Driver Aggression Score'], 2)}, and tire usage aggression was "
        f"{round(row['Tire Usage Aggression'], 2)}. "
        f"The lap time variation for this driver was {round(row['Lap Time Variation'], 6)}. "
    )

    # Weather
    summary += (
        f"Weather conditions included an air temperature of {row['Air_Temp_C']}°C, track temperature of "
        f"{row['Track_Temp_C']}°C, humidity at {row['Humidity_%']}%, and wind speeds of {row['Wind_Speed_KMH']} km/h."
    )

    return summary
```

```python
f1_clean['Summary'] = f1_clean.apply(generate_summary, axis=1)

f1_clean.head()
```

```python
# Merge all stints into one summary
f1_grouped = f1_clean.groupby(['Season', 'Race Name', 'Driver'])['Summary'] \
               .apply(lambda x: ' '.join(x)) \
               .reset_index(name='CombinedSummary')

f1_grouped
```

# Chunking

```python
import spacy
nlp = spacy.load("en_core_web_md")

def sentence_chunking(text, sentences_per_chunk):
    doc = nlp(text)
    sentences = [sent.text.strip() for sent in doc.sents]

    chunks = []
    for i in range(0, len(sentences), sentences_per_chunk):
        chunk = " ".join(sentences[i:i + sentences_per_chunk])
        chunks.append(chunk)

    return chunks
```

```python
sentences_per_chunk = 1 

# Add chunks column
f1_grouped['Chunks'] = f1_grouped['CombinedSummary'].apply(
    lambda text: sentence_chunking(text, sentences_per_chunk)
)
```

```python
f1_grouped[['Driver', 'Race Name', 'Chunks']].head(3)
```

# Embedding

```python
f1_chunks_df = f1_grouped.explode('Chunks').reset_index(drop=True)
f1_chunks_df = f1_chunks_df.rename(columns={'Chunks': 'ChunkText'})
```

```python
# Using spaCy for prototype
texts = f1_chunks_df['ChunkText'].tolist()  # list of strings
docs = list(nlp.pipe(texts, batch_size=32))  # fast batch processing

f1_chunks_spacy = f1_chunks_df.copy()
f1_chunks_spacy['embedding'] = [doc.vector for doc in docs]  # extract embeddings
```
